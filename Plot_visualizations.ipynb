```python
import matplotlib.pyplot as plt

bit_sizes = [64, 256, 512, 1024]
###CPU metrics
cpu_custom = [0.000030, 0.000139, 0.000413, 0.002647]
cpu_bignum = [0.000031, 0.000033, 0.000099, 0.000643]
##GPU CuPY metrics
gpu_cupy_encrypt = [0.011848, 0.031335, 0.066196]
gpu_cupy_decrypt = [0.027494, 0.075907, 0.157805]
gpu_cupy_bits = [64, 128, 192]
##Pure CUDA
pure_cuda_encrypt = [0.0001275]
pure_cuda_decrypt = [0.0001623]
pure_cuda_bits = [64]

## Comparing for CPU only
plt.figure(figsize=(10, 5))
plt.plot(bit_sizes, cpu_custom, marker='o', label='CPU Custom Barrett')
plt.plot(bit_sizes, cpu_bignum, marker='s', label='CPU OpenSSL BIGNUM')
plt.xlabel("Bit ")
plt.ylabel(" Time (s)")
plt.title("CPU RSA Implementation")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#  GPU Comparisonn
plt.figure(figsize=(10, 5))
plt.plot(gpu_cupy_bits, gpu_cupy_encrypt, marker='o', label=' CuPy Encrypt')
plt.plot(gpu_cupy_bits, gpu_cupy_decrypt, marker='s', label=' CuPy Decrypt')
plt.plot(pure_cuda_bits, pure_cuda_encrypt, marker='^', label=' CUDA Encrypt')
plt.plot(pure_cuda_bits, pure_cuda_decrypt, marker='v', label=' CUDA Decrypt')
plt.xlabel("Bit  / Limbs ")
plt.ylabel(" Time (s)")
plt.title("GPU RSA Modular Implementation")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# CPU / GPU 
plt.figure(figsize=(8, 5))
labels = ['CPU Barrett', 'CPU BIGNUM', 'GPU CuPy E', 'GPU CuPy D', 'Pure CUDA E', 'Pure CUDA D']
values = [0.000030, 0.000031, 0.011848, 0.027494, 0.0001275, 0.0001623]
plt.bar(labels, values, color=['b', 'g', 'r', 'orange', 'purple', 'brown'])
plt.ylabel("Average Time (s)")
plt.title("64-bit RSA Implentation")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
